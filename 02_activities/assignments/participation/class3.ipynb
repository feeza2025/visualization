```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy 
import PIL
import requests
```

```python
# Adding for Reproducibility
np.random.seed(613)

# Create random values for X and Y
x = np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
# Create the figure and plot (the plot is called our axes)
fig, ax  = plt.subplots(figsize=(5,3))
ax.plot(x, y1, label = "Person 1")
ax.plot(x, y2, label = "Person 2")
ax.legend(loc='lower right',
          frameon = True, #add frame around the legend
          fontsize = 12, #change font size
          ncol = 2, #specify number of columns
          shadow = True,
          )
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left', bbox_to_anchor =(1, 1))
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!", 
        ha='center', ##alignment of text 
        color = 'red', ##modify font colour
        size = 20)
fig.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2, color='green')
font1 = {'family': 'sans-serif', 'color': 'blue', 'size':14}
font2 = {'family': 'monospace', 'color': 'green', 'size':14}
# add title to the plot
ax1.set_title('AX 1', fontdict = font1)
ax1.set_xlabel('number', fontdict = font1)

ax2.set_title('AX 2', fontdict = font2)
ax2.set_xlabel('character', fontdict = font2)

fig.show()
```

```python
plt.style.use('default')
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax1'], ['ax2', 'ax3']], figsize=(7, 4))
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
plt.show()
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], ['ax2', 'ax3']], figsize=(7, 4), layout = "constrained")
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
#fig.tight_layout()
```

```python
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images

response = requests.get('https://static.wikia.nocookie.net/the-ultimate-good/images/3/3d/Luffy.png/revision/latest?cb=20230417200546')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
```

```python
ax_image = fig.add_axes([0.1, # x coordinate (ON FIGURE, NOT AXES) 
                         0.11, # y coordinate (ON FIGURE, NOT AXES) 
                         0.15, # image width 
                         0.35] # image height
                         )
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax_image = fig.add_axes([0.1, 0.11, 0.15, 0.35])

ax_image.imshow(image)
ax_image.axis('off')
```
