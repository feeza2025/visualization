```python
tips = sns.load_dataset("tips")
print(tips)
```

```python
sns.set_style('whitegrid')
sns.lineplot(data=tips, # choose our dataset 
             x='total_bill', # define our x variable 
             y='tip') # define our y variable
```

```python
tipgraph = sns.lineplot(data=tips, 
                        x='total_bill', 
                        y='tip', 
                        color = 'hotpink', 
                        linestyle = '--', 
                        linewidth = 3, 
                        marker = 'o', 
                        markerfacecolor = 'indigo')
```

```python
tipgraph = sns.scatterplot(data=tips, #dataset
                           x='total_bill', 
                           y='tip', 
                           style = 'time', # change style of our marker according to the time variable
                           hue = 'day', # change hue of our marker according to the day variable
                           palette = ['purple', 'hotpink', 'deepskyblue', 'yellowgreen']) # colors - will use default if not specified
tipgraph.set(title='Tips vs. Total Bill', 
             xlabel='Total Bill ($)',
             ylabel='Tip Amount ($)')
```

```python
# compare potential relationships very efficiently
# takes each numeric dataset and pair against each other - useful for exploring relationship
# the area is where a variable is pair against itself
sns.pairplot( data = tips, 
             hue = 'day')
```

```python
daysplot = sns.relplot( data=tips, # dataset
                       x="total_bill", # x axis title
                       y="tip", # y axis title
                       hue="sex", # cdifferent color dots to represent sex
                       col="day", # separate each graph by day
                       kind="scatter", # type of graph
                       col_wrap=2) # put graphs in 2 columns
```

```python
import plotly.graph_objects as go # 'go' is 'graph objects’
import nbformat
```

```python
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
# make our figure object
graph = go.Figure()
```

```python
# choose our plot type(bar) and our data
graph.add_trace(go.Bar(x=x1, y=y1))

# Next, update our layout to include titles
graph.update_layout( title="Pirate Scores", 
                    xaxis_title="Pirates", 
                    yaxis_title="Score")

# show our plot
graph.show()
```

```python
graph.write_html("pirategraph.html")
```

```python
graph = go.Figure()
graph.add_trace(go.Scatter(x=x1, 
                           y=y1, 
                           mode='markers', # we want points for a scatter plot 
                           marker=dict( # what they say is a dictionary - not python dictionary 
                               size=15, # point size 
                               color='hotpink', # point colour 
                               opacity=1, # point transparency/alpha 
                               line=dict(width=5, color='purple') # point outline 
                               )))

graph.update_layout( 
    title='Interactive Pirate Plot', 
    xaxis_title='Pirates', 
    yaxis_title='Scores', 
    width=500, # pixels
    height=500)
```

```python
from wordcloud import WordCloud
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",on_bad_lines='skip')
df
```

```python
# join all our text from each row from our quote column into a string
text = " ".join(each for each in df.quote)
# # generate our wordcloud image
wordcloud = WordCloud(background_color="white",
                      colormap = 'inferno',
                      random_state= 613).generate(text)

# use matplotlib syntax to put our image in a figure

fig, ax = plt.subplots(figsize=(7, 3))

ax.imshow(wordcloud) # remember 'imshow' from when we added pictures to our matplotlib axes 
          #interpolation='bilinear') # this line helps smooth our image
ax.axis("off")
```

```python
# import the packages
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
# define our sets
A = set(["apple", "banana", "watermelon"]) # fruits
B = set(["pumpkin", "blueberry", "apple", "key lime"]) # pies
```

```python
diagram = venn2([A, B], set_labels = ('Fruits', 'Pies'), set_colors=("blue", "red"), alpha=0.5) # alpha is transparency

# customizing
# separate elements by newline
# the 10, 11, 01 are like gates - 1 reps on - the two switches are the circles
diagram.get_label_by_id("10") .set_text("\n".join(A - B)) # A only
diagram.get_label_by_id("11") .set_text("\n".join(A & B)) # A intersect B
diagram.get_label_by_id("01") .set_text("\n".join(B - A)) # B only

plt.show()
```
