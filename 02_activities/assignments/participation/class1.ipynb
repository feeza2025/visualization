```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
fig, ax  = plt.subplots(figsize=(5,3))
ax.scatter(x, y)
```

```python
fig, ax  = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
fig, ax  = plt.subplots(figsize=(5,3))
ax.plot(x,y)
```

```python
fig, ax  = plt.subplots(figsize=(5,3))
ax.hist(y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
fig, ax  = plt.subplots(figsize=(5,3))
ax.plot(y)

font1 = {'family': 'sans-serif', 'color': 'blue', 'size':20}
font2 = {'family': 'monospace', 'color': 'green', 'size':14}

ax.set_title('Total growth over time', fontdict = font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
fig.tight_layout()

```

```python
fig, ax  = plt.subplots(figsize=(5,3))
ax.scatter(x,
           y,
           marker='*',
           color='indigo')
fig.show()
```

```python
fig, ax  = plt.subplots(figsize=(5,3))
ax.plot(x,
           y,
           marker='*',
           color='indigo',
           linestyle='--',
           linewidth=2)
fig.show()
```

```python
fig, ax  = plt.subplots(figsize=(5,3))
ax.plot(x,
           y,
           marker='*',
           color='#ff0000',
           linestyle='--',
           linewidth=2,
           markersize=12,
           markeredgecolor= '#0f0',
           markerfacecolor= '#fff',)
ax.grid(axis='y', color='#f00', linewidth=2, linestyle='-.')
fig.show()
```
